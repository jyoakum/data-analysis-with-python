```python
import matplotlib.pyplot as plt
from numpy.random import randn
```

```python
baseline = []
for x in range(1000):
  baseline.append(2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
_ = ax.plot(baseline, 'k--')
# The 'k--' is a style option instructing matplotlib to plot a black dashed line.
```

```python
_ = ax.plot(randn(1000).cumsum())
```

```python
_ = ax.plot(randn(1000).cumsum(), '--')
# The '--' is a style option instructing matplotlib to plot a dashed line.
```

```python
x_ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=0, fontsize='small')
```

```python
_ = ax.set_title('My Two Plots')
```

```python
_ = ax.set_xlabel('Stages')
```
